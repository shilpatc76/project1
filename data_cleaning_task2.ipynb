```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("tested.csv")
```

```python
print(df.head())
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nStatistical Summary:")
print(df.describe(include='all'))
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin (too many missing values)
df.drop('Cabin', axis=1, inplace=True)

print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df, palette='pastel')
plt.title("Survival Count (0 = Not Survived, 1 = Survived)")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df, palette='coolwarm')
plt.title("Survival Rate by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Pclass', y='Survived', data=df, palette='muted')
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(x='Survived', y='Age', data=df, palette='pastel')
plt.title("Age vs Survival")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("\n INSIGHTS:")
print("1 Females had a much higher survival rate than males.")
print("2 Passengers in 1st class had better survival chances.")
print("3 Younger passengers tended to survive more than older ones.")
print("4 Higher fare correlated positively with survival chances.")
```
