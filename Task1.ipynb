```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

# Define age bins and labels
bins = [20, 30, 40, 50, 60, 100]
labels = ["20-30", "31-40", "41-50", "51-60", "60+"]

# Create age groups
df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)

# Group by AgeGroup and Occupation (counts)
age_occ = df.groupby(["AgeGroup", "Occupation"]).size().unstack(fill_value=0)

# Plot stacked bar chart with counts
plt.figure(figsize=(12, 7))
age_occ.plot(kind="bar", stacked=True, figsize=(12,7), colormap="tab20")

plt.title("Occupation by Age Group (Counts)")
plt.xlabel("Age Group")
plt.ylabel("Number of People")
plt.xticks(rotation=0)
plt.legend(title="Occupation", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
